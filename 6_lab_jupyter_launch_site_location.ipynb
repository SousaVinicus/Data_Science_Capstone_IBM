```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

After you plot distance lines to the proximities, you can answer the following questions easily:

*   Are launch sites in close proximity to railways?
      * Sim. Os locais de lançamento estão estrategicamente posicionados próximos a ferrovias para facilitar o transporte logístico de componentes pesados e volumosos do foguete (como os estágios do Falcon 9) das fábricas até o local de integração.
*   Are launch sites in close proximity to highways?
      * Sim. A proximidade com rodovias é essencial para o acesso rápido de pessoal, suprimentos diários e equipamentos menores necessários para as operações de pré-lançamento.
*   Are launch sites in close proximity to coastline?
      * Sim. Esta é a característica geográfica mais crítica. Estar na costa permite que a trajetória de lançamento ocorra sobre o oceano, minimizando o risco para áreas habitadas em caso de falha e permitindo que os estágios descartados caiam com segurança na água.
*   Do launch sites keep certain distance away from cities?
      * Sim. Os locais são mantidos a uma distância considerável de grandes centros urbanos por razões de segurança pública (risco de explosões) e para mitigar o impacto da poluição sonora extrema gerada pelos motores durante a decolagem.
